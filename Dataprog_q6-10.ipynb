```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
file_path = 'C:/Users/hp/Downloads/us-marriages-divorces-1867-2014.csv'
data = pd.read_csv(file_path)

# Assuming the CSV file has columns 'Year', 'Marriages', and 'Divorces'
years = data['Year']
marriages_per_capita = data['Marriages']
divorces_per_capita = data['Divorces']

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(years, marriages_per_capita, label='Marriages per Capita', color='blue')
plt.plot(years, divorces_per_capita, label='Divorces per Capita', color='red')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Number per Capita')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')

# Adding legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
file_path = 'C:/Users/hp/Downloads/us-marriages-divorces-1867-2014.csv'
data = pd.read_csv(file_path)

# Filter data for the years 1900, 1950, and 2000
selected_years = [1900, 1950, 2000]
selected_data = data[data['Year'].isin(selected_years)]

# Assuming the CSV file has columns 'Year', 'Marriages', and 'Divorces'
years = selected_data['Year']
marriages_per_capita = selected_data['Marriages']
divorces_per_capita = selected_data['Divorces']

# Plotting
plt.figure(figsize=(10, 6))

bar_width = 0.35
bar_positions_marriages = [pos - bar_width/2 for pos in range(len(selected_years))]
bar_positions_divorces = [pos + bar_width/2 for pos in range(len(selected_years))]

plt.bar(bar_positions_marriages, marriages_per_capita, width=bar_width, label='Marriages per Capita', color='blue')
plt.bar(bar_positions_divorces, divorces_per_capita, width=bar_width, label='Divorces per Capita', color='red')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Number per Capita')
plt.title('Comparison of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')

# Adding legend
plt.legend()

# Set x-axis ticks and labels
plt.xticks(range(len(selected_years)), selected_years)

# Display the plot
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
 
file_path = r'C:\Users\hp\Downloads\actor_kill_counts.csv'
data = pd.read_csv(file_path)
 
sorted_data = data.sort_values(by='Count', ascending=True)
 
plt.figure(figsize=(10, 6))
plt.barh(sorted_data['Actor'], sorted_data['Count'], color='orange')
 
# Adding labels and title
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood by Kill Count')
 
# Label each bar
for index, value in enumerate(sorted_data['Count']):
    plt.text(value, index, str(value))
 
# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
 
file_path = r'C:\Users\HP\Downloads\roman-emperor-reigns.csv'
data = pd.read_csv(file_path)
assassination_count = data['Cause_of_Death'].str.contains('assassinat', case=False).sum()
 
total_emperors = len(data)
fraction_assassinated = assassination_count / total_emperors
fraction_not_assassinated = 1 - fraction_assassinated
 
# Create labels
labels = ['Assassinated', 'Not Assassinated']
sizes = [fraction_assassinated, fraction_not_assassinated]
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Fraction of Roman Emperors Assassinated')
 
# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the CSV file
file_path = 'C:/Users/hp/Downloads/arcade-revenue-vs-cs-doctorates.csv'
data = pd.read_csv(file_path)

# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total Arcade Revenue (billions)', y='Computer Science Doctorates Awarded (US)', hue='Year', data=data, palette='viridis')

# Adding labels and title
plt.xlabel('Total Arcade Revenue (billions)')
plt.ylabel('Number of Computer Science PhDs Awarded (US)')
plt.title('Relationship between Arcade Revenue and CS PhDs (2000-2009)')

# Display the plot
plt.legend(title='Year')
plt.show()
```
